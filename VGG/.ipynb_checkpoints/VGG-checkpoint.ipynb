```python
#load VGG Model
from keras.applications.vgg16 import VGG16
help(VGG16)
```

```python
#load model
model = VGG16(weights='vgg_weight/vgg16_weights_tf_dim_ordering_tf_kernels.h5')
```

```python
# load and prepare image
from keras.preprocessing.image import load_img
# load an image 
image = load_img('test.jpg', target_size=(224, 224))
```

```python
image
```

```python
# keras work with Numpy array ,so we need to convert the image
from keras.preprocessing.image import img_to_array
image = img_to_array(image)
```

```python
"""
The network expects one or more images as input
that means the input array will need to be 4-dimensional
: samples, rows, columns, and channels.
"""
# we only have one
image = image.reshape((1,image.shape[0],image.shape[1],image.shape[2]))
```

```python
'''
the image pixels need to be prepared in the same way as
the ImageNet training data was prepared. 
'''
from keras.applications.vgg16 import preprocess_input
image = preprocess_input(image)
```

```python
# make a prediction
yhat = model.predict(image)
```

```python
# we need to interpret the probabilities
from keras.applications.vgg16 import decode_predictions
# convert the prob to class labels
label = decode_predictions(yhat)
label
```

```python
name = label[:,1]
```

```python
name = []
for x in label:
    for y in x:
        name.append(y[1]) 
```

```python
import http.client
import hashlib
from urllib import parse
import random

def Baidu_Translate(text):
    
    appid = '20180811000193465' #你的appid
    secretKey = 'RWu6IqRarQzAzwTrbzwF' #你的密钥

 
    httpClient = None
    myurl = '/api/trans/vip/translate'
    q = text # 翻译的句子
    fromLang = 'en'
    toLang = 'zh'
    salt = random.randint(32768, 65536)
    
    sign = appid+q+str(salt)+secretKey
    m1 = hashlib.md5()
    m1.update(sign.encode(encoding='utf-8'))
    sign = m1.hexdigest()
    myurl = myurl+'?appid='+appid+'&q='+parse.quote(q)+'&!from='+fromLang+'&to='+toLang+'&salt='+str(salt)+'&sign='+sign
 
    try:
        httpClient = http.client.HTTPConnection('api.fanyi.baidu.com')
        httpClient.request('GET', myurl)
 
        #response是HTTPResponse对象
        response = httpClient.getresponse()
        str1 = response.read().decode('utf-8')
        str1 = eval(str1)
        return str1['trans_result'][0]['dst']
    except Exception as e:
        print(e)
    finally:
        if httpClient:
            httpClient.close()
```
