```python
poetry_file = 'data/poetry.txt'

poetry_list = [] # 唐诗存放数组

with open(poetry_file, 'r', encoding='utf-8', errors='ignore') as f:
    f_lines = f.readlines()
    print('唐诗总数:{}'.format(len(f_lines)))
    # 逐行处理
    for line in f_lines:
        # 去除前后空白符
        strip_line = line.strip()
        try:
            # 将唐诗分为标题和内容
            title, content = strip_line.split(':')
        except:
            # 出现多个':'的将被舍弃
            continue
        # 去除内容中的空格
        content = content.strip().replace(' ','')
        # 舍弃含有非法字符的唐诗
        if '(' in content or '（' in content or '<' in content or '《' in content or '_' in content or '[' in content:
            continue
        # 舍弃过短或过长的唐诗
        lenth = len(content)
        if lenth < 20 or lenth > 100:
            continue
        # 加入列表
        poetry_list.append('s' + content + 'e')
```

### 唐诗 ==> 词汇表

```python
'''
list.append(object) 向列表中添加一个对象object
list.extend(sequence) 把一个序列seq的内容添加到列表中
'''
```

```python
words_list = []
# 获取唐诗中所有的字符
for poetry in poetry_list:
    words_list.extend([word for word in poetry])

# 统计每个字出现次数    
counter = collections.Counter(words_list)
#降序排序
sorted_word = sorted(counter.items(),key=lambda x:x[1],reverse=True)
# 解压 字 , 次数
words_list, _ = zip(*sorted_word)
words_list = words_list[:len(words_list)]+(' ',)
print('词汇表大小:{}'.format(len(words_list)))
words_list
```

### 生成词汇向量

```python
# 生成单词到ID 的映射,每个字对应一个数字ID
word_id_dict = dict(zip(words_list, range(len(words_list))))
# 将poetry_list转换成向量形式
to_num = lambda word:word_id_dict.get(word,len(words_list))
poetry_vector=[list(map(to_num, poetry)) for poetry in poetry_list]

# 将向量写入文件
VECTOR_DATA = 'data/poetry_vector_data.txt'
with open(VECTOR_DATA,'w') as f:
    for id_l in poetry_vector:
        f.write(''.join(str(i) for i in id_l)+'\n')

poetry_vector
```

```python
# 每次取64首进行训练
batch_size = 64
n_chunk = len(poetry_vector) // batch_size #5//4=1

class DataSet(object):
    def __init__(self, data_size):
        self._data_size = data_size #64
        self._epochs_completed = 0
        self._index_in_epoch = 0
        self._data_index = np.arange(data_size)# 0,1,2,3-63
        
    # 获取下一个batch
    def next_batch(self, batch_size):
        start = self._index_in_epoch #第几代
        if start + batch_size > self._data_size:
            np.random.shuffle(self._data_index) # 打乱顺序函数
            self._epochs_completed += 1 # 完成了几代
            self._index_in_epoch = batch_size 
            full_batch_features, full_batch_labels = self.data_batch(0, batch_size)
            return full_batch_features,full_batch_labels
        else:
            self._index_in_epoch += batch_size
            end = self._index_in_epoch
            full_batch_features,full_batch_labels = self.data_batch(start,end)
            if self._index_in_epoch == self._data_size:
                self._index_in_epoch = 0
                self._epochs_completed += 1
                np.random.shuffle(self._data_index)
            return full_batch_features,full_batch_labels
        
    def data_batch(self, start, end):
        batches = [] #一个batche 的诗集数字向量
        for i in range(start, end):
            batches.append(poetry_vector[self._data_index[i]])
            
        
        length = max(map(len,batches)) # 最长的句子作为列
        #填充 end - start = batch_size
        xdata = np.full((end - start, length),word_id_dict[' '],np.int32)
        # 赋值 
        for row in range(end - start):
            xdata[row,:len(batches[row])] = batches[row]
        #仅相当于填充
        ydata = np.copy(xdata)
        #从XDATA的第二个开始赋值
        ydata[:,:-1] = xdata[:,1:]
        """
        xdata             ydata
        [6,2,4,6,9]       [2,4,6,9,9]
        [1,4,2,8,5]       [4,2,8,5,5]
        """
        
        return xdata, ydata
```

# RNN 模型

```python
len(words_list)
tf.reset_default_graph()  
```

```python
input_data = tf.placeholder(tf.int32,[batch_size,None])
output_targets = tf.placeholder(tf.int32,[batch_size,None])

#定义RNN
def neural_network(model='lstm',rnn_size=128,num_layers=2):
    if model == 'rnn':
        cell_fun = tf.nn.rnn_cell.BasicRNNCell
    elif model == 'gru':
        cell_fun = tf.nn.rnn_cell.GRUCell
    elif model == 'lstm':
        cell_fun = tf.nn.rnn_cell.BasicLSTMCell
    
    # lstm cell is divided into two parts (c_state,m_state)
    cell = cell_fun(rnn_size, state_is_tuple=True)
    cell = tf.nn.rnn_cell.MultiRNNCell([cell]*num_layers, state_is_tuple=True)
    
    # 初始化state 为0
    initial_state = cell.zero_state(batch_size, tf.float32)
    tf.get_default_graph
    with tf.variable_scope('rnnlm', reuse=tf.AUTO_REUSE): #共享变量模块 名为"rnnlm"
        # 创建或获取变量
        softmax_w = tf.get_variable("softmax_w",[rnn_size, len(words_list)])
        softmax_b = tf.get_variable("softmax_b",[len(words_list)])
        with tf.device("/cpu:0"):
            embedding = tf.get_variable("embedding",[len(words_list),rnn_size])
            inputs = tf.nn.embedding_lookup(embedding, input_data)
        
        # last_State=(c_state, m_State)
        outputs, last_state = tf.nn.dynamic_rnn(cell, inputs
                                               , initial_state=initial_state
                                              , scope='rnnlm')
        
        # 上面返回的outputs是一个tensor,所以要化成一个list
        # output 是所有的Y  
        output = tf.reshape(outputs, [-1, rnn_size])
        # matmul 二维计算
        # Y
        logits = tf.matmul(output, softmax_w) + softmax_b
        probs = tf.nn.softmax(logits) #预测的句子
        return logits, last_state, probs, cell, initial_state

# 加载已有模型
def load_model(sess, saver, ckpt_path):
    latest_ckpt = tf.train.latest_checkpoint(ckpt_path)
    if latest_ckpt:
        print("resum from", latest_ckpt)
        saver.restore(sess, latest_ckpt)
        return int(latest_ckpt[latest_ckpt.rindex('-')+1:])
    else:
        print("building new model")
        sess.run(tf.global_variables_initializer())
        return -1
```

```python
def train_neural_network():
    logits, last_state, _, _, _ = neural_network()
    targets = tf.reshape(output_targets, [-1]) # 化成一行向量
    #?
    '''
      loss = tf.contrib.legacy_seq2seq.sequence_loss_by_example(
            [logits],  # 预测的结果
            [tf.reshape(self.targets, [-1])],  # 期待的正确答案，这里讲[batch_size,num_steps]二维数组压缩成一维数组
            [tf.ones([batch_size * num_steps], dtype=tf.float32)]  # 损失的权重，在这里所有的权重都为1，也就是说不同batch和不同时刻的重要程度是一样的
            )
    '''
    loss = tf.contrib.legacy_seq2seq.sequence_loss_by_example([logits]
                                                  , [targets]
                                                  , [tf.ones_like(targets, dtype=tf.float32)]
                                                  , len(words_list))
    cost = tf.reduce_mean(loss)
    learning_rate = tf.Variable(0.0, trainable=False)
    # 获取全部需要 训练的变量
    tvars = tf.trainable_variables()
    """
    让权重的更新限制在一个合适的范围。
    tf.clip_by_global_norm(t_list, clip_norm, use_norm=None, name=None)
    """ 
    grads, _ = tf.clip_by_global_norm(tf.gradients(cost, tvars),5)
    # 创建一个优化器，设置他的学习率
    optimizer = tf.train.AdamOptimizer(learning_rate)
    # 创建一个 梯度下降 operation
    train_op = optimizer.apply_gradients(zip(grads,tvars))
    
    # 配置 session中设备的使用
    session_config = tf.ConfigProto(allow_soft_placement = True)
    # 按需求增加GPU使用率
    session_config.gpu_options.allow_growth = True
    
    
    trainDS = DataSet(len(poetry_vector))
    
    with tf.Session(config=session_config) as sess:
        sess.run(tf.initialize_all_variables())
        saver = tf.train.Saver(tf.all_variables())
        
        
        for epoch in range(5):
            # 学习率衰退
            sess.run(tf.assign(learning_rate,0.002 * (0.97 ** epoch)))
            all_loss = 0.0
            for batche in range(n_chunk):
                x,y = trainDS.next_batch(batch_size)
                train_loss, _, _= sess.run([cost, last_state, train_op],
                                           feed_dict={input_data:x, output_targets:y})
                all_loss += train_loss
                if batche % 50 == 1:
                    print(epoch, batche, 0.002 * (0.97 ** epoch),train_loss)
                    
            saver.save(sess,'model/poetry.module',global_step=epoch)
            print(epoch,'LOSS:',all_loss * 1.0 / n_chunk)
```
