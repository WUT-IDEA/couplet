```python
import collections
import numpy as np
import tensorflow as tf
import time
```

```python
poetry_file = 'data/poetry.txt'

poetry_list = [] # 唐诗存放数组

with open(poetry_file, 'r', encoding='utf-8', errors='ignore') as f:
    f_lines = f.readlines()
    print('唐诗总数:{}'.format(len(f_lines)))
    # 逐行处理
    for line in f_lines:
        # 去除前后空白符
        strip_line = line.strip()
        try:
            # 将唐诗分为标题和内容
            title, content = strip_line.split(':')
        except:
            # 出现多个':'的将被舍弃
            continue
        # 去除内容中的空格
        content = content.strip().replace(' ','')
        # 舍弃含有非法字符的唐诗
        if '(' in content or '（' in content or '<' in content or '《' in content or '_' in content or '[' in content:
            continue
        # 舍弃过短或过长的唐诗
        lenth = len(content)
        if lenth < 20 or lenth > 100:
            continue
        # 加入列表
        poetry_list.append('[' + content + ']')
```

```python
# 按字数排序
poetry_list = sorted(poetry_list,key=lambda x:len(x))
poetry_list
```

```python
words_list = []
# 获取唐诗中所有的字符
for poetry in poetry_list:
    words_list.extend([word for word in poetry])

# 统计每个字出现次数    
counter = collections.Counter(words_list)
#降序排序
sorted_word = sorted(counter.items(),key=lambda x:x[1],reverse=True)
# 解压 字 , 次数
words_list, _ = zip(*sorted_word)
words_list = words_list[:len(words_list)]+(' ',)
print('词汇表大小:{}'.format(len(words_list)))
words_list
```

```python
# 生成单词到ID 的映射,每个字对应一个数字ID
word_id_dict = dict(zip(words_list, range(len(words_list))))
# 将poetry_list转换成向量形式
to_num = lambda word:word_id_dict.get(word,len(words_list))
poetry_vector=[list(map(to_num, poetry)) for poetry in poetry_list]
poetry_vector
```

```python
# 每次取64首进行训练
batch_size = 1
n_chunk = len(poetry_vector) // batch_size
x_batches = []
y_batches = []
for i in range(n_chunk):
    start_index = i * batch_size
    end_index = start_index + batch_size

    batches = poetry_vector[start_index:end_index]
    length = max(map(len,batches))
    xdata = np.full((batch_size,length), word_id_dict[' '], np.int32)
    for row in range(batch_size):
        xdata[row,:len(batches[row])] = batches[row]
    ydata = np.copy(xdata)
    ydata[:,:-1] = xdata[:,1:]
    """
    xdata             ydata
    [6,2,4,6,9]       [2,4,6,9,9]
    [1,4,2,8,5]       [4,2,8,5,5]
    """
    x_batches.append(xdata)
    y_batches.append(ydata)
```

```python
input_data = tf.placeholder(tf.int32,[batch_size,None])
output_targets = tf.placeholder(tf.int32,[batch_size,None])

#定义RNN
def neural_network(model='lstm',rnn_size=128,num_layers=2):
    if model == 'rnn':
        cell_fun = tf.nn.rnn_cell.BasicRNNCell
    elif model == 'gru':
        cell_fun = tf.nn.rnn_cell.GRUCell
    elif model == 'lstm':
        cell_fun = tf.nn.rnn_cell.BasicLSTMCell
    
    # lstm cell is divided into two parts (c_state,m_state)
    cell = cell_fun(rnn_size, state_is_tuple=True)
    cell = tf.nn.rnn_cell.MultiRNNCell([cell]*num_layers, state_is_tuple=True)
    
    # 初始化state 为0
    initial_state = cell.zero_state(batch_size, tf.float32)
    
    with tf.variable_scope('rnnlm',reuse=tf.AUTO_REUSE): #共享变量模块 名为"rnnlm"
        # 创建或获取变量
        softmax_w = tf.get_variable("softmax_w",[rnn_size, len(words_list)])
        softmax_b = tf.get_variable("softmax_b",[len(words_list)])
        embedding = tf.get_variable("embedding",[len(words_list),rnn_size])
        inputs = tf.nn.embedding_lookup(embedding, input_data)
        
        # last_State=(c_state, m_State)
        outputs, last_state = tf.nn.dynamic_rnn(cell, inputs
                                               , initial_state=initial_state
                                              , scope='rnnlm')
        
        # 上面返回的outputs是一个tensor,所以要化成一个list
        # output 是所有的Y  
        output = tf.reshape(outputs, [-1, rnn_size])
        # matmul 二维计算
        # Y
        logits = tf.matmul(output, softmax_w) + softmax_b
        probs = tf.nn.softmax(logits) #预测的句子
        return logits, last_state, probs, cell, initial_state
```

```python
def gen_poetry_with_head(head,type):
    if type !=5 and type != 7:
        print("必须是5个字或者7个字一句")
        return
    
    def to_word(weights):
        # ？
        t = np.cumsum(weights)
        s = np.sum(weights)
        sample = int(np.searchsorted(t, np.random.rand(1)*s))
        return words_list[sample]
    
    # laststate 是一个LSTMStatetube，probs 是tensor，cell ，initial_state是tensor
    _, last_state, probs, cell, initial_state = neural_network()  # 返回的是tensor。即节点即操作！！
  
    with tf.Session() as sess:
        sess.run(tf.initialize_all_variables())
        
        saver = tf.train.Saver(tf.all_variables())
        saver.restore(sess, 'model/poetry.module-4')
        with tf.variable_scope('rnnlm',reuse=True): #共享变量模块 名为"rnnlm"
            # 创建或获取变量
            softmax_w = tf.get_variable("softmax_w",[128, len(words_list)])
            print(sess.run(softmax_w))
        state_ = sess.run(cell.zero_state(1, tf.float32))
        
        poem = ''
        i = 0
        for the_word in head:
                flag = True
                while flag:
                    state_ = sess.run(cell.zero_state(1, tf.float32))
                    x = np.array([list(map(word_id_dict.get, '['))])
                    print("1input_data,x:",x)
                    [probs_, state_] = sess.run([probs, last_state], feed_dict={input_data: x, initial_state: state_})
                    sentence = the_word
                    print("1sentence:",sentence)
                    x = np.zeros((1, 1))
                    x[0, 0] = word_id_dict[sentence]
                    print("2input_data,x:",x)
                    [probs_, state_] = sess.run([probs, last_state], feed_dict={input_data: x, initial_state: state_})
                    word = to_word(probs_)
                    sentence += word
                    print("2sentence",sentence)
                    while word!='。':
                        x = np.zeros((1, 1))
                        x[0, 0] = word_id_dict[word]
                        [probs_, state_] = sess.run([probs, last_state], feed_dict={input_data: x, initial_state: state_})
                        word = to_word(probs_)
                        print("循环里的word",word)
                        sentence += word
                        print("循环里的sentence",sentence)
                        if len(sentence) == 2 + 2 * type:
                            sentence += '\n'
                            poem += sentence
                            flag = False
        return poem
```
